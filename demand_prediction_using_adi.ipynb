```python
GB_RETAILS['InvoiceDate']=pd.to_datetime(GB_RETAILS['InvoiceDate'])

```

```python
GB_RETAILS['day']=GB_RETAILS['InvoiceDate'].dt.dayofweek
```

```python
GB_RETAILS['dateformat']=GB_RETAILS['InvoiceDate'].dt.strftime('%Y-%m-%d')
```

```python
GB_RETAILS['dateformat']=pd.to_datetime(GB_RETAILS['dateformat'])
```

```python
GB_RETAILS['dateformat'].value_counts()
```

```python
GB_GROUPED=GB_RETAILS.groupby(['Description','dateformat']).agg(total_sale=('Quantity','sum')).reset_index()
```

```python
GB_GROUPED.head()
```

```python
#calculating avg and standard deviation

cvdata=GB_GROUPED.groupby('Description').agg(average=('total_sale','mean'), sd=('total_sale','std')).reset_index()
```

```python
#calculating cv_sq
cvdata['cv_square']=(cvdata['sd']/cvdata['average'])**2
```

```python
cvdata.head()
```

```python
#adi per product

prod_by_date=GB_RETAILS.groupby(['Description','dateformat']).agg(count=('Description','count')).reset_index()
```

```python
prod_by_date.head(40)
```

```python
sku=prod_by_date.Description.value_counts()
```

```python
sku.head(30)
```

```python
new_df=pd.DataFrame()
```

```python
for i in range(len(sku.index)):
    a=prod_by_date[prod_by_date['Description']==sku.index[i]]
    
    a['previos date']=a['dateformat'].shift(1)
    new_df=pd.concat([new_df,a],axis=0)
    
    
```

```python
new_df.info()
```

```python
new_df['duration']=new_df['dateformat']- new_df['previos date']
```

```python
new_df['dateformat']=pd.to_datetime(new_df['dateformat'])
new_df['previos date']=pd.to_datetime(new_df['previos date'])
```

```python
new_df.info()
```

```python
new_df['duration']=new_df['duration'].astype(str).str.replace('days','')
```

```python
#u need to convert timedelta to numeric to find the adi value
```

```python
new_df['duration']=pd.to_numeric(new_df['duration'],errors='coerce')
```

```python
#cal adi

ADI=new_df.groupby('Description').agg(ADI=('duration','mean')).reset_index()

```

```python
ADI.head(100)
```

```python
#
adicv=pd.merge(ADI,cvdata)
```

```python
adicv.head()
```

```python
def category(df):
    a=0
    
    if((df['ADI']<=1.34) & (df['cv_square']<=0.49)):
        a='Smooth'
    if((df['ADI']>=1.34) & (df['cv_square']>=0.49)):  
        a='Lumpy'
    if((df['ADI']<1.34) & (df['cv_square']>0.49)):
        a='Erratic'
    if((df['ADI']>1.34) & (df['cv_square']<0.49)):
        a='Intermittent'
    return a
```

```python

#categorizing products based on forecastability
adicv['category']=adicv.apply(category,axis=1)
```
