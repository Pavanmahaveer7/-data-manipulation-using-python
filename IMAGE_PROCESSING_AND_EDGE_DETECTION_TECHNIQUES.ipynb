```python
import cv2

import numpy as np

import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow

```

```python
image=cv2.imread("/content/albert.jpg")
```

#sobel edge detection:An image processing technique called edge detection is used to find the edges (or boundaries of objects or regions in an image. Edges are one of the most important aspects of images.


```python
sobeledge_x=cv2.Sobel(image, cv2.CV_64F,0,1,ksize=3)
sobeledge_y=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)

```

```python
cv2_imshow(image)
cv2_imshow(sobeledge_x)

cv2_imshow(sobeledge_y)


```

```python
OR_IMAGE=cv2.bitwise_or(sobeledge_x,sobeledge_y)

cv2_imshow(OR_IMAGE)

cv2.waitKey(0)
```

```python
import cv2
import numpy as np
img2=cv2.imread("/content/albert.jpg")


Kernalx=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
Kernaly=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
PrewittX=cv2.filter2D(img2,-1,Kernalx)
PrewittY=cv2.filter2D(img2,-1,Kernaly)
PRE=cv2.add(PrewittX,PrewittY)


cv2_imshow(img2)
cv2_imshow(PrewittX)
cv2_imshow(PrewittY)
cv2_imshow(PRE)
cv2.waitKey(0)
```

#The Prewitt operator detects image edges by convolution with two filter masks. One for horizontal and one for vertical direction.Prewitt operator convolution with two filter masks detects image edges. vertical and horizontal directions, respectively.

```python
import cv2

logimg=cv2.imread("/content/albert.jpg")

logdetec=cv2.Laplacian(logimg,-1)
cannyimg=cv2.Canny(logimg,80,120)
#double thresholding
cv2_imshow(logdetec)
#cv2_imshow(cannyimg)
#we need give threshold for canny for tracking edges

cv2.waitKey(0)
```

# Canny edge detector is a multi-step algorithm to detect the edges for any input image. 

```python
import cv2

logimg=cv2.imread("/content/albert.jpg")

logdetec=cv2.Laplacian(logimg,-1)
cannyimg=cv2.Canny(logimg,80,120)
#double thresholding
#cv2_imshow(logdetec)
cv2_imshow(cannyimg)
#we need give threshold for canny for tracking edges

cv2.waitKey(0)
```
